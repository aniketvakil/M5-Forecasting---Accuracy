```python
submission1 = pd.read_csv('../input/killeravg/submissionliller.csv')
submission2 = pd.read_csv('../input/dark-magic/full 1280.csv')
submission3 = pd.read_csv('../input/dark-magic/full 1300.csv')
submission4 = pd.read_csv('../input/dark-magic/full 1350.csv')
submission5 = pd.read_csv('../input/bestst/best.csv')
```

```python
submission1 = submission1.sort_values(by = 'id').reset_index(drop = True)
submission2  = submission2.sort_values(by = 'id').reset_index(drop = True)
submission3 = submission3.sort_values(by = 'id').reset_index(drop = True)
submission4  = submission4.sort_values(by = 'id').reset_index(drop = True)
submission5 = submission5.sort_values(by = 'id').reset_index(drop = True)
sub5 = submission1.copy()

for i in sub5.columns :
    if i != 'id' :
        sub5[i] = 0.02*submission1[i] + 0.03*submission2[i]+0.03*submission3[i] + 0.02*submission4[i]+ 0.9*submission5[i]
        
sub5.to_csv('submissionf.csv', index = False)
```
